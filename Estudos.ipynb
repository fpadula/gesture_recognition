### Análise da distribuição das classes

```python
import pandas as pd

header = []
for id_no in range(21):
    for coord in ["x", "y"]:
        header.append(f"ID_{id_no}_{coord}")

header = ["Gesture_ID"] + header
keypoints = pd.read_csv("./hand-gesture-recognition-mediapipe/model/keypoint_classifier/keypoint.csv", 
                        names=header)

keypoints.head()
```

```python
gesture_labels = pd.read_csv("./hand-gesture-recognition-mediapipe/model/keypoint_classifier/keypoint_classifier_label.csv",
                    header=None)
gesture_labels = gesture_labels.values.tolist()
gesture_labels = [label[0] for label in gesture_labels]
gesture_labels
```

```python
from matplotlib import pyplot as plt
import numpy as np


fig = plt.figure(figsize=(8,6))
ax = fig.gca()
labels, counts = np.unique(keypoints["Gesture_ID"], return_counts=True)

plt.bar(gesture_labels, counts, align='center')
ax.set_xticks(labels);
# fig = plt.grid(True)
```
